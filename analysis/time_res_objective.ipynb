```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.use("pgf")
plt.rcParams['pgf.texsystem'] = 'pdflatex'
plt.rcParams['pgf.rcfonts'] = False
plt.rcParams['figure.edgecolor'] = 'k'
plt.rcParams['figure.facecolor'] = 'w'
plt.rcParams['savefig.dpi'] = 400
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = "serif"
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df.index, df.LC, marker='h', label='Least Cost', color='tab:green', markersize=12, lw=2)
ax.plot(df.index, df.XN, marker='^', label='Expensive Nuclear', color='tab:blue', markersize=12, lw=2)
ax.plot(df.index, df.ZAN, marker='x', label='Zero Advanced Nuclear', color='gray', markersize=12, lw=2)
ax.plot(df.index, df.ZN, marker='s', label='Nuclear Phaseout', color='yellow', markersize=12, lw=2)

labels = {0:'Seasonal', 1:'Monthly', 2:'Weekly', 3:'Daily'}
plt.xticks(list(labels.keys()), list(labels.values()), fontsize=14)
plt.yticks(fontsize=14)
ax.set_xlabel('Time Resolution', fontsize=20)
ax.set_ylabel('Temporal Resolution', fontsize=20)
ax.set_ylabel(r'System Levelized Cost of Electricity [\$/MWh]', fontsize=20)
ax.legend(fontsize=20, shadow=True, fancybox=True)
ax.set_title('Influence of Temporal Resolution\n on System Cost', fontsize=32)
plt.savefig('../thesis/figures/time_res_cost_plot.pgf')
# plt.show()
```
