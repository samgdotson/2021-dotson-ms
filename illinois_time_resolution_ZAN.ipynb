```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.use("pgf")
plt.rcParams['pgf.texsystem'] = 'pdflatex'
plt.rcParams['pgf.rcfonts'] = False
plt.rcParams['figure.edgecolor'] = 'k'
plt.rcParams['figure.facecolor'] = 'w'
plt.rcParams['savefig.dpi'] = 400
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = "serif"
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,20), sharex=True)

w=0.1

# Capacity

(cap_df_4/1000).plot.bar(ax=axes[0,0], stacked=True, 
                         color=elc_colors, width=w, 
                         position=1.2, align='center',
                         legend=False,edgecolor='gray')
(cap_df_12/1000).plot.bar(ax=axes[0,0], stacked=True, 
                          color=elc_colors, width=w, 
                          position=0, align='center',
                          legend=False,edgecolor='gray')
(cap_df_52/1000).plot(ax=axes[0,0], kind='bar', stacked=True, 
                      color=elc_colors, width=w, 
                      position=-1.2, align='center',
                      legend=False, edgecolor='gray')


# Generation
(gen_df_4/1000).plot.bar(ax=axes[0,1], stacked=True, 
                         color=elc_colors, width=w, 
                         position=1.2, align='center',
                         legend=False,edgecolor='gray')
(gen_df_12/1000).plot.bar(ax=axes[0,1], stacked=True, 
                          color=elc_colors, width=w, 
                          position=0, align='center',
                          legend=False,edgecolor='gray')
(gen_df_52/1000).plot(ax=axes[0,1], kind='bar', stacked=True, 
                      color=elc_colors, width=w, 
                      position=-1.2, align='center',
                      legend=False, edgecolor='gray')


(stacked_cap_4).plot.bar(ax=axes[1,0], stacked=True, 
                         color=elc_colors, width=w, 
                         position=1.2, align='center',
                         legend=False,edgecolor='gray')
(stacked_cap_12).plot.bar(ax=axes[1,0], stacked=True, 
                          color=elc_colors, width=w, 
                          position=0, align='center',
                          legend=False,edgecolor='gray')
(stacked_cap_52).plot(ax=axes[1,0], kind='bar', stacked=True, 
                      color=elc_colors, width=w, 
                      position=-1.2, align='center',
                      legend=False, edgecolor='gray')


(stacked_gen_4).plot.bar(ax=axes[1,1], stacked=True, 
                         color=elc_colors, width=w, 
                         position=1.2, align='center',
                         legend=False,edgecolor='gray')
(stacked_gen_12).plot.bar(ax=axes[1,1], stacked=True, 
                          color=elc_colors, width=w, 
                          position=0, align='center',
                          legend=False,edgecolor='gray')
(stacked_gen_52).plot(ax=axes[1,1], kind='bar', stacked=True, 
                      color=elc_colors, width=w, 
                      position=-1.2, align='center',
                      legend=False, edgecolor='gray')

axes[0,0].set_ylabel("[GW]", fontsize=20)
axes[0,1].set_ylabel("[TWh]", fontsize=20)
axes[1,0].set_ylabel("[%]", fontsize=20)
axes[1,1].set_ylabel("[%]", fontsize=20)
# axes[1].set_ylabel("Installed Capacity [%]", fontsize=20)


axes[1,0].set_xlabel("Year", fontsize=20)
axes[1,1].set_xlabel("Year", fontsize=20)
axes[1,0].tick_params(axis='x', labelsize=16, rotation=0)
axes[1,1].tick_params(axis='x', labelsize=16, rotation=0)

axes[0,0].tick_params(axis='y', labelsize=16)
axes[0,1].tick_params(axis='y', labelsize=16)
axes[1,0].tick_params(axis='y', labelsize=16)
axes[1,1].tick_params(axis='y', labelsize=16)

axes[0,0].set_title("Installed Capacity", fontsize=24)
axes[0,1].set_title("Total Generation", fontsize=24)

handles, labels = axes[0,1].get_legend_handles_labels()
by_label = dict(zip(labels, handles))
legend = plt.figlegend(by_label.values(), by_label.keys(), bbox_to_anchor=(0.99,0.01), ncol=7, 
              fontsize=16, shadow=True, title='Technologies')
plt.suptitle('Illinois: 2030 Net Zero Electricity at 3 Time Resolutions \n Scenario: Expensive/Zero Advanced Nuclear \n', 
             fontsize=32)
plt.setp(legend.get_title(),fontsize=24)
plt.tight_layout()
plt.savefig('thesis/figures/time_resolution_plot_ZAN.pgf')
```
