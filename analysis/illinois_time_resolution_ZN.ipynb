```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import pygenesys as pg
from plotter import *
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,12)
```

```python
def get_objective(conn):
    """
    Retrieves the objective function
    """
    cursor = conn.cursor()
    command = "SELECT * FROM Output_Objective"
    cursor.execute(command)
    obj_value = cursor.fetchall()[0][-1]
    
    return obj_value
```

```python
sim_path = "/home/sdotson/research/2021-dotson-ms/simulations/illinois/zero_nuclear/IL_ZN_CC30_4.sqlite"
conn = sqlite3.connect(sim_path)
cursor = conn.cursor()

# plot_electricity_profiles(conn,'IL','ELC')
cap_df_4 = get_annual_capacity(conn, 'IL', 'ELC' )
(cap_df_4/1000).plot(kind='bar', stacked=True, color=elc_colors)
plt.title('Capacity Plot for Scenario: Business As Usual')
plt.ylabel('Capacity [GW]')
plt.minorticks_on()
plt.show()
gen_df_4 = get_annual_generation(conn, 'IL', 'ELC' )
(gen_df_4).plot(kind='bar', stacked=True, color=elc_colors)
plt.ylabel('Generation [GWh]')
plt.minorticks_on()
plt.show()
```
